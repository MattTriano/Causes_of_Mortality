```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
cm_all_ages_df = pd.read_csv('WHO_mortality_data\ghe2016_deaths_country_allages.csv')
country_conv = pd.read_csv('WHO_mortality_data\iso3_ctry_codes.csv')
populations = pd.read_csv('WHO_mortality_data\worldbank_population_data.csv', na_values=['..'])
country_conv = country_conv.drop('m49_code', axis=1)
populations.tail(6)
```

From this printout of the last 6 rows of population data, we see that the last 5 rows are just fluff. We can remove them.

```python
populations.drop(populations.tail(5).index, inplace=True)
populations.tail(6)
```

```python
populations = populations.drop(['Series Code',  'Series Name'], axis=1)
populations.rename(columns=lambda x: x.replace('20', 'pop20'), inplace=True)
populations.rename(columns=lambda x: x.split()[0] if 'pop' in x else x, inplace=True)
```

* Years starting with 'dths' represent the estimated number of deaths for that year. 
* Years starting with 'low' or 'upp" represent the lower or uppe bound on the estimate for that year. (we can probably trim these)
* iso3: This represents the country
* cause2015: This is a numeric value that maps 1:1 to the 'causename' column

Data for this visualization was from http://terrance.who.int/mediacentre/data/ghe/

Population Data from Worldbook http://databank.worldbank.org/data/reports.aspx?source=2&series=SP.POP.TOTL#

From the printout above, we see that Zimbabwe is the last country, and all of the populations after that are aggregated. We want populations for just the countries.

```python
zimbabwe = populations.index[populations['Country Name'] == 'Zimbabwe'][0]
populations = populations[:zimbabwe+1]
```

```python
def missing_values(df, name):
    print("Missing values in " + str(name))
    print(df.isnull().sum())
```

Ok, we see that population values are missing just for Eritrea, and from from 2012 onward. I can find and input these values. According to this [UN population tracker](http://data.un.org/Data.aspx?q=population&d=PopDiv&f=variableID%3a12), the median annual population values for Eritrea are as below.

```python
populations.loc[(populations['Country Code'] == 'ERI'), 'pop2012'] = 4560977
populations.loc[(populations['Country Code'] == 'ERI'), 'pop2013'] = 4650998
populations.loc[(populations['Country Code'] == 'ERI'), 'pop2014'] = 4746045
populations.loc[(populations['Country Code'] == 'ERI'), 'pop2015'] = 4846976
populations.loc[(populations['Country Code'] == 'ERI'), 'pop2016'] = 4954645
populations.loc[(populations['Country Code'] == 'ERI'), 'pop2017'] = 5068831
```

```python
mort_df = pd.merge(left=country_conv, right=cm_all_ages_df, on='iso3', how='inner')
cm_filter = ['iso3', 'country_name', 'cause2015', 'causename', 'sex']   
df_base = mort_df[cm_filter]
dths_filter = [col for col in mort_df if col.startswith('dths')]
cm_filter.extend(dths_filter)
mort_dths_df = mort_df[cm_filter]
mort_pop_dths_df = pd.merge(left=mort_dths_df, right=populations, left_on='iso3', right_on='Country Code', how='inner')
mort_pop_dths_df.head()
```

```python
pop_cols = [x for x in mort_pop_dths_df.columns if 'pop' in x]
for pop_col in pop_cols:
    mort_pop_dths_df[pop_col] = mort_pop_dths_df[pop_col].astype(int)
```

```python
mort_pop_dths_df.info()
```

Ok, now that I've merged both the population data and the death counts, I can normalize the death counts to population.

```python
# start_dths = mort_pop_dths_df.columns.get_loc('dths2000')
# start_pops = mort_pop_dths_df.columns.get_loc('pop2000')
mort_per_capita = df_base
for i in range(2000, 2017):
    dths = str('dths' + str(i))
    pops = str('pop' + str(i))
    mort_per_capita[str(i)] = mort_pop_dths_df[dths]*100000/mort_pop_dths_df[pops]
```

```python
mort_per_capita['causename'] = mort_per_capita['causename'].str.replace(',', ' ')
mort_per_capita['causename'] = mort_per_capita['causename'].str.replace('  ', ' ')
mort_per_capita['causename'] = mort_per_capita['causename'].str.replace(' ', '_')
```

```python
mort_per_capita = mort_per_capita[mort_per_capita['sex'] == 'BTSX']
mort_per_capita.drop('sex', axis=1, inplace=True)
```

```python
mort_per_capita['avg_death_rate'] = (mort_per_capita['2000'] + mort_per_capita['2001'] + mort_per_capita['2002'] + 
                                     mort_per_capita['2003'] + mort_per_capita['2004'] + mort_per_capita['2005'] +
                                     mort_per_capita['2006'] + mort_per_capita['2007'] + mort_per_capita['2008'] +
                                     mort_per_capita['2009'] + mort_per_capita['2010'] + mort_per_capita['2011'] +
                                     mort_per_capita['2012'] + mort_per_capita['2013'] + mort_per_capita['2014'] +
                                     mort_per_capita['2015'] + mort_per_capita['2016'])/17
```
